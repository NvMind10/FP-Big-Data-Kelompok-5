```python
from pymongo import MongoClient

import os
from dotenv import load_dotenv, find_dotenv

load_dotenv(find_dotenv())
MONGO_URI = os.getenv("MONGO_URI")

client = MongoClient(MONGO_URI)

db = client["scamshield"]

collection = db["telegram_logs"]

print("Database ScamShield siap")
```

```python
from datetime import datetime

sample_data = {
    "timestamp": datetime.now().isoformat(),
    "sender": "@redred",
    "group": "INFO LOKER BOGOR 2026",
    "message_type": "image",
    "message_text": "Kerja dari rumah cukup like youtube",
    "status_ai": "SCAM",
    "confidence_score": 98.41
}

result = collection.insert_one(sample_data)

print("Inserted ID:", result.inserted_id)
```

```python
collection.create_index("sender")
collection.create_index("status_ai")
collection.create_index("timestamp")

print("Index berhasil dibuat")
```

CELL 6

```python
import pandas as pd

df1 = pd.read_csv(
    "part-00000-0ef21029-d701-403f-b148-1c18cfe5780c-c000.csv",
    header=None,
    names=["sender","text","status_ai"]
)

df2 = pd.read_csv(
    "part-00000-0ce57249-d7e0-484d-9401-ca1a4aba8d09-c000.csv",
    header=None,
    names=["sender","text","status_ai"]
)

df3 = pd.read_csv(
    "part-00000-00aca37f-8364-4abc-83b3-fba6fcde1f4a-c000.csv",
    header=None,
    names=["sender","text","status_ai"]
)

df = pd.concat([df1, df2, df3], ignore_index=True)

print("Jumlah Data:", len(df))

df.head()
```

CELL 7

```python
from datetime import datetime

records = []

for _, row in df.iterrows():

    record = {
        "timestamp": datetime.now().isoformat(),
        "sender": str(row["sender"]),
        "group": "telegram_group",
        "message_type": "text",
        "message_text": str(row["text"]),
        "status_ai": str(row["status_ai"]),
        "confidence_score": None,
        "source": "spark_pipeline"
    }

    records.append(record)

print("Records siap:", len(records))
```

CELL 8

```python
result = collection.insert_many(records)

print("Inserted:", len(result.inserted_ids))
```

```python
pipeline = [
    {
        "$group": {
            "_id": "$sender",
            "jumlah": {"$sum": 1}
        }
    },
    {
        "$sort": {"jumlah": -1}
    },
    {
        "$limit": 10
    }
]

for x in collection.aggregate(pipeline):
    print(x)
```
